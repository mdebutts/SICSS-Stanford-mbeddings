```python
import requests
from bs4 import BeautifulSoup
import csv
import time

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

def get_news_urls(filename="tkww_news_urls.csv"):
    base_url = "https://www.tkww.hk/top_news/more_{}.html"
    seen_urls = set()
    page_number = 1  # Start at 1 instead of 10000
    consecutive_empty_pages = 0  # Stop when no new URLs appear
    
    print("Starting URL collection...")
    
    with open(filename, "a", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["url"])
    
        while consecutive_empty_pages < 3:  # Allow a few empty pages before stopping
            url = base_url.format(page_number)
            print(f"Fetching: {url}")
            
            for attempt in range(3):  # Retry up to 3 times
                try:
                    response = requests.get(url, headers=headers, timeout=10)
                    response.raise_for_status()
                    break  # If successful, exit retry loop
                except requests.exceptions.RequestException as e:
                    print(f"Error fetching {url} (attempt {attempt + 1}/3): {e}")
                    time.sleep(5)  # Wait before retrying
            else:
                print(f"Skipping {url} after 3 failed attempts.")
                page_number += 1
                continue  # Move to next page
            
            soup = BeautifulSoup(response.text, 'html.parser')
            new_urls = set()
            
            for link in soup.find_all('a', class_='common-column-list-unit-title'):
                article_url = link.get('href')
                if article_url and article_url.startswith("https://www.tkww.hk/a/"):
                    new_urls.add(article_url)
            
            if new_urls - seen_urls:
                print(f"Found {len(new_urls - seen_urls)} new URLs on page {page_number}")
                new_entries = new_urls - seen_urls
                seen_urls.update(new_entries)
                
                for url in new_entries:
                    writer.writerow([url])
                    f.flush()
                
                consecutive_empty_pages = 0  # Reset counter
            else:
                consecutive_empty_pages += 1
                print(f"No new URLs on page {page_number} (empty streak: {consecutive_empty_pages})")
            
            page_number += 1
            time.sleep(1)  # Be polite to the server
    
    print(f"Finished scraping. {len(seen_urls)} total URLs collected.")
    return list(seen_urls)

def main():
    get_news_urls()
    print("Scraping complete.")
```

```python
import requests
from bs4 import BeautifulSoup
import csv
import time
import re

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

def read_urls_from_csv(filename="tkww_news_urls.csv"):
    urls = []
    with open(filename, "r", encoding="utf-8") as f:
        reader = csv.reader(f)
        next(reader)  # Skip header
        for row in reader:
            urls.append(row[0])
    return urls

def clean_publication_time(time_str):
    return re.sub(r" HKT", "", time_str.strip())

def scrape_article(url):
    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
    except requests.exceptions.RequestException as e:
        print(f"Failed to fetch {url}: {e}")
        return None
    
    soup = BeautifulSoup(response.text, 'html.parser')
    title_tag = soup.find('h1', class_='content-title-text')
    time_tag = soup.find('span', class_='content-title-tag-right')
    content_div = soup.find('div', class_='content-body-inner')
    
    title = title_tag.text.strip() if title_tag else ""
    publication_time = clean_publication_time(time_tag.text) if time_tag else ""
    content = " ".join([p.text.strip() for p in content_div.find_all('p')]) if content_div else ""
    
    return {
        'url': url, 
        'title': title, 
        'publication_time': publication_time, 
        'content': content
    }

def save_article_to_csv(article, filename="tkww_news_articles.csv"):
    file_exists = False
    try:
        with open(filename, "r", encoding="utf-8") as f:
            file_exists = True
    except FileNotFoundError:
        pass
    
    with open(filename, "a", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["url", "title", "publication_time", "content"])
        if not file_exists:
            writer.writeheader()
        writer.writerow(article)
        print(f"Saved article: {article['title']}")

def main():
    urls = read_urls_from_csv()
    
    for i, url in enumerate(urls, 1):
        print(f"Scraping article {i}/{len(urls)}: {url}")
        article_data = scrape_article(url)
        if article_data:
            save_article_to_csv(article_data)
        time.sleep(1)  # Be polite
    
    print("Scraping complete.")

main()
```
